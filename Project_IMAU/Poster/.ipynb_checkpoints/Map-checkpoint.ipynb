```python
import numpy as np
import scipy as sp
import xarray as xr
import json
import matplotlib.pyplot as plt
import cartopy

%matplotlib ipympl 
# for interactive plots
import pandas as pd 
# for removing NaNs from data
import pyproj 
# for projecting lat/long to x/y (cartopy may have an API for this but I already know pyproj)
```

Download dan van Blackboard de file `SODrifters.json`, zet die in dezelfde directory, map of folder als je notebook en laad de data met de code hieronder

```python
def reshape_jsonarrays(item):
    time = np.array([d[0] for d in item[1]], dtype='datetime64[ns]')
    lat = np.array([d[1] for d in item[1]], dtype='float')
    lon = np.array([d[2] for d in item[1]], dtype='float')
    return {'name':item[0], 'time':time, 'lat':lat, 'lon':lon}

with open("SOdrifters.json") as fp:
    jsondata = json.load(fp)

drifters = [reshape_jsonarrays(item) for item in jsondata.items()]
print(f"Aantal drifters: {len(drifters)}")
```

Je hebt nu een lijst van 20 drifter trajectories. Maak nu een heel simpele plot van longitude versus latitude van de trajecten van alle drifters met `plt.plot()`

```python
plt.figure('drifters')
for drifter in drifters:
    plt.plot(drifter['lon'], drifter['lat'])
plt.title(f"Paden van de {len(drifters)} drifters")
plt.grid()
plt.show()
```

We zien hierboven het traject van de drifters. Ter orientatie: de drifters zijn rond 155 $^\circ$ W te water gelaten, en bewegen van daar langzaam naar het oosten. Zoals je ziet is het een behoorlijke spaghetti van lijnen!

Voor wat meer context over de paden kan het handig zijn ook de continenten te plotten. Hiervoor gebruiken we het `cartopy` package, bijvoorbeeld met de code hieronder. 

*Let op dat het, afhankelijk van de versie van je cartopy module, kan zijn dat je een aantal `warnings` ziet. Dat is niet erg, die waarschuwingen kun je negeren.*

```python
# We gebruiken hieronder de PlateCarree projectie;
# op de Cartopy website kun je ook andere projecties vinden
# als je daar meet wilt spelen
projection = cartopy.crs.PlateCarree

# initialisatie van de figuur
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(1, 1, 1, projection=projection(central_longitude=180))

# plot de kustlijnen en maak het land beige
ax.coastlines(resolution='50m')
ax.add_feature(cartopy.feature.LAND)

# plot de trajecten, let op dat we een transformatie moeten gebruiken
for drifter in drifters:
    plt.plot(drifter['lon'], drifter['lat'], transform=projection())

# zoom out om meer land te zien
ax.set_extent((142,228,-77,-32), crs=projection())

# de code hieronder plot lichtgrijze lijnen voor het longitude/latitude grid
# en zorgt voor goede formatting van de labels
gl = ax.gridlines(crs=projection(), draw_labels=True, linewidth=0.5,
                  color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xformatter = cartopy.mpl.gridliner.LONGITUDE_FORMATTER
gl.yformatter = cartopy.mpl.gridliner.LATITUDE_FORMATTER
plt.show()
```

Zoals je zult zien zit er flink wat 'drift' (een quasi-lineaire trend) in de data. Dit komt door de stroming van de oceaan. Om de inertiaalbewegingen te vinden, moeten we deze drift eruit halen. Dit kun je bijvoorbeeld doen door de lineaire trend van de data af te trekken. Dit kun je doen met de `scipy` module, bijvoorbeeld met `scipy.signal.detrend()`.

<h4>Opgave 2.2</h4> [10 punten]: Plot nu de breedtegraad van de drifter (in alleen het deel dat je in opgave 2.1 extra goed weer hebt gegeven) zonder de lineaire drift als functie van tijd. Kun je nu de inertiaalbewegingen herkennen?

```python
lat_detrend = sp.signal.detrend(interest['lat'][740:1860])
time_short = interest['time'][740:1860]

plt.figure('detrend')
plt.plot(time_short, lat_detrend)
plt.ylabel('Difference in latitude to trend')
plt.xlabel('Date')
plt.show()
```

<h5>Antwoord 2.2</h5> In bovenstaande grafiek zijn 23 pieken te zien, en de horizontale as heeft een lengte van 47 dagen of 4,060,800 seconden. In dat geval is de frequentie $5.66\cdot10^{-6}\mathrm{~~s^{-1}}$.</br>
Dit zal ik nemen als schatting voor de frequentie: deze zal rond $10^{-6} \mathrm{~~s^{-1}}$ zal zitten.

Waarschijnlijk zie je nu (naast wellicht wat laag-frequentie variatie) een duidelijke sinusvorm in de data. Om de frequentie van deze sinusvorm te vinden, kun je bijvoorbeeld een fourieranalyse doen. Dit kun je het makkelijkst doen met de `plt.magnitude_spectrum()` functie van `matplotlib`.

<h4>Opgave 2.3</h4> [10 punten]: Plot de frequentie-spectrum van de (gedetrende) breedtegraad van de drifter en bereken hieruit de frequentie van de inertiaalbewegingen.

```python
output = np.abs(np.fft.rfft(lat_detrend))
frequencies = np.fft.rfftfreq(len(lat_detrend), np.mean(np.diff(time_short)/np.timedelta64(1, 's')))

peaks,_ = sp.signal.find_peaks(output, height=20, distance=5)
peak = peaks[-1]

plt.figure('fourier')
plt.plot(frequencies, output, label='Fourier decomposition of the signal')
height = output[peak]
frequency_dat = frequencies[peak]
plt.vlines(frequency_dat, 0, height, 'k', '--', label='Highest frequency peak above intensity 20')
plt.ylabel('Magnitude')
plt.xlabel('Frequency (s$^{—1}$)')
plt.legend()
plt.show()
print(f'Highest frequency: {frequency_dat}')
```

<h5>Antwoord 2.3</h5> Ik deel door een timedelta van 1 seconde om de eenheid te laten kloppen. Hierna gebruik ik scipy om de pieken te vinden. Zodra ik alle pieken heb neem ik degene met de hoogste frequentie (ik hoef me door de minimale height en distance geen zorgen te maken om ruis). Dit zal de piek zijn van de inertiaalbewegingen.

<h3>Opgave 2.4</h3> [5 punten]: Hoe groot is het verschil met de theoretische waarde?

```python
f = lambda lat, Omega=7.292e-5 : 2*Omega*np.sin(lat)

coriolis = np.mean(f(interest['lat'][740:1860]))
frequency_the = np.reciprocal(2*np.pi/np.abs(coriolis))

print(f'Theoretical value given by reciprocal of the inertial period: {frequency_the} s-1')
print(f'Hence, the difference between the observed and theoretical values is {frequency_dat - frequency_the} s-1 (factor {frequency_dat/frequency_the})')
```

Merk op: ik gebruik de gemiddelde Corioliskracht op de breedtegraden, niet de Corioliskracht op de gemiddelde breedtegraad.</p>

<h4>Opgave 3.1</h4> [10 punten]: Schrijf nu zelf een functie die de snelheid van de drifter met nummer 130258 (de eerste in de lijst: `drifters[0]`) uitrekent. Let op dat je de locatie in graden noorderbreedte en oosterlengte geeft, en dat de snelheid in meter per seconde moet zijn. Je moet verschillen in graden omrekenen naar meters.

```python
total_latitude = interest['lat']
total_longitude = interest['lon']


for i,(lat,lon) in enumerate(zip(total_latitude, total_longitude)):
    if lat>180: total_latitude[i] -= 360
    if lon>180: total_longitude[i] -= 360

transformer = pyproj.Transformer.from_crs("WGS84", "EPSG:6167")

xs, ys = np.asarray(transformer.transform(total_latitude, total_longitude))

earth_radius = 40_075.017e3
for i,x in enumerate(xs):
    if x<0: xs[i] += earth_radius

positions = np.vstack((xs, ys)).T

positions -= positions[0]

positions = positions.T

velocity = np.hypot(*np.gradient(positions, axis=1))

plt.figure('speed_xy')
plt.plot(interest['time'], velocity)
plt.xlabel('Date')
plt.ylabel(r'Speed (m s$^{-1}$)')
plt.show()

print(f'The total eventual displacement according to this projection is {np.hypot(*positions)[-1]:.4e} m.')
```

<h5>Antwoord 3.1</h5> 
<p>De code neemt de coordinaten van de drifter en brengt deze eerst in het domein $[-180^\circ,180^\circ]$. Hierna worden deze met behulp van de <a href='https://epsg.io/6167-datum'>EPSG:6167-projectie</a> omgezet tot locaties in meters, die daarna gemoduleerd worden zodat deze niet van groot positief naar groot negatief gaan (het omslagpunt zit bij 180$^\circ$ O/W). Hierna wordt, om de snelheden te bepalen, de gradiënt van deze posities over tijd genomen.</p>
<p>Volgens Google Earth is de afstand tussen de start- en eindcoördinaten $\approx3.51887\cdot10^{6}~\mathrm{m}$. Het verschil tussen deze en de gevonden waarde is erg klein, met een factor 29/30. Deze projectie komt wat betreft locaties dus overeen met Google Earth. Hierom denk ik dat deze transformatie in dit geval goed is.</p>

<h5>Antwoord 3.3</h5>
Indien er een perfecte correlatie is, zal deze matrix de $\left(\begin{smallmatrix}1&1\\1&1\end{smallmatrix}\right)$ matrix zijn. Deze matrix ligt er redelijk dicht bij (de niet-diagonale elementen zijn in de orde $10^{-1}$), wat aanduidt dat het model redelijk dicht bij de werkelijkheid ligt.

Dit omdat beide onbekende onzekerheden hebben.</p>

Een mogelijke reden waarom het model en de observaties verschillen kan zijn dat het model dagelijkse data geeft, terwijl de drifter data elk uur is. Om dit te onderzoeken kun je de snelheid van de drifter middelen over een dag. 

<h4>Opgave 3.4</h4> [10 punten]: Schrijf een functie die voor elke dag de gemiddelde snelheid van de drifter berekent in zowel de observaties en het model, plot wederom de zonale snelheden en bereken de correlaties van deze dag-gemiddelden. Wat valt je op?

```python
def dayAvg(us, poss, tims):
    outsU = {}
    outsP = {}
    times = []

    for i,j,k in zip(us, tims, poss):
        times.append(np.datetime64(str(j)[:10]))
        try:
            outsU[str(j)[:10]].append(np.float64(i))
            outsP[str(j)[:10]].append(np.float64(k))
        except KeyError:
            outsU[str(j)[:10]] = [np.float64(i)]
            outsP[str(j)[:10]] = [np.float64(k)]

    averageU = np.asarray([np.mean(np.float64(outsU[i]), axis=0) for i in outsU])
    averageP = np.asarray([np.mean(np.float64(outsP[i]), axis=0) for i in outsP])
    return averageU, averageP, np.unique(times)


uObsAvg, xObsAvg, tObsAvg = dayAvg(uObs_prune, np.asarray([interest['lat'],interest['lon']]).T, time_prune)
uModDay = ds_uv.uo.interp(time=('z', tObsAvg),
                         longitude=('z', xObsAvg.T[1]),
                         latitude=('z', xObsAvg.T[0])).isel(depth=0)

fig, ax1 = plt.subplots()
ax2 = plt.twinx()
driftline, = ax1.plot(tObsAvg, uObsAvg, 'b--', label='Drifter')
modelline, = ax2.plot(tObsAvg, uModDay, 'r-', label='Model')
ax1.set_ylabel('Zonale driftersnelheid')
ax2.set_ylabel('Zonale modelsnelheid')

# Syncing the 0 points of both axes, done in such a way that the left axis is unchanged and the right axis matches.
mL, ML = ax1.get_ylim()
mR, MR = ax2.get_ylim()

wL = abs(mL)/(ML+abs(mL))
wR = abs(mR)/(MR+abs(mR))

MR = (mR*ML/mL if wL<=wR else MR      )
mR = (mR       if wL<=wR else MR*mL/ML)

ax1.set_ylim(mL,ML)
ax2.set_ylim(mR,MR)


plt.legend([modelline, driftline], ['Model', 'Data'])
plt.show()

print(f'Correlation matrix: \n{np.corrcoef(uObsAvg, uModDay)}')
```

<h5>Antwoord 3.4</h5>
Deze correlatiecoëfficiëntmatrix is verder weg van de $\left(\begin{smallmatrix}1&1\\1&1\end{smallmatrix}\right)$ matrix dan degene in opdracht (3.3). Dit duidt aan dat er mogelijk een fout zit in de berekening. Echter hebben we de modeldata opnieuw geïnterpoleerd en de geobserveerde data slechts gemiddeld,  dus de oorzaak hiervan is onbekend.</br>
In het plaatje zien we dat er nog steeds een aantal ordegroottes verschil is tussen de modeldata en de drifterdata. We zien ook nog steeds de horizontale verschuiving. Ik verwacht dat deze dezelfde oorzaak hebben als voorheen.</br>
Verder zien we in de functies ook een verschil rond eind januari/begin februari 2014. Dit is ook waar de inertiaalbewegingen het meest zichtbaar waren. Deze zouden mogelijk de oorzaak kunnen zijn: het model lijkt een hogere versnelling te verwachten rond 21 januari, waarna de traagheid die hogere gemodellerde snelheid door laat lopen tot het eind van die inertiaalbewegingen.</br>

Tot slot gaan we kijken naar de zeespiegelvaritie langs _alle_ drifterpaden. Hiervoor hebben we een file `eta.nc` op blackboard gezet.

<h4>Opgave 4.1</h4> [5 punten]: Schrijf een functie die de zeespiegelvariatie langs een drifterpad berekent. Plot de zeespiegelvariatie als functie van de tijd voor alle drifters.

```python
eta = xr.open_dataset('eta.nc')
eta
```

```python
def path_variation(drifter):
    datLon = drifter['lon']
    datLat = drifter['lat']
    datTim = drifter['time']
    result = eta.zos.interp(time=('z', datTim),
                         longitude=('z', datLon),
                         latitude=('z', datLat)).isel()

    return result

seaLevelsAll = [path_variation(i) for i in drifters]
```

```python
plt.figure('sea_levels')
for i,j in zip(seaLevelsAll,drifters):
    plt.plot(j['time'], i)
plt.xlabel('Date')
plt.ylabel('Sea Level Difference')
plt.show()
```

<h5>Antwoord 4.1</h5>
<p>Bonvenstaande grafiek toont de zeespiegel over tijd op alle drifterpaden. Het is te zien dat, hoewel alle paden erg chaotisch zijn, ze allemaal een trend naar boven hebben.</p>

<h4>Opgave 4.2</h4> [10 punten]: Wat valt je op aan de plot die je hierboven hebt gemaakt? Hoe kun je de langzame trend in de waarden verklaren, met behulp van de theorieën die je in AOD hebt geleerd?

<h5>Antwoord 4.2</h5>
Elk van de lijnen in de bovenstaande figuur gaat langzaam omhoog. Dit kan met barotrope stroom te maken hebben: de Coriolisconstante $f$ op deze breedtegraad is negatief dus \begin{equation}\mathrm{sign~}{\mathbf{u}_g}=\mathrm{sign~}{\left(\frac{\partial\eta}{\partial y}\right)}\end{equation} \begin{equation}\mathrm{sign~}{\mathbf{v}_g}=-\mathrm{sign~}\left(\frac{\partial\eta}{\partial x}\right)\end{equation}
waarbij $\mathbf{u}_g$ de zonale snelheid en $\mathbf{v}_g$ de meridionale snelheid.</br>
Eerdere figuren tonen dat de drifters naar het oosten gaan. Hieruit volgt dus dat $\frac{\partial\eta}{\partial y}$ positief moet zijn. Dit wordt ook ondersteund door het tijdsgemiddelde zeeniveau dat hieronder geplot is. In dezelfde plot zien we ook dat het zeeniveau op elke individuele breedtegraad nog veel verandert. Dit verklaart waarom er weinig noord-zuid beweging te zien is: de $\mathbf{v}_g$ middelt zichzelf ongeveer uit. </br>
Dus de stijgende trend in zeespiegel over tijd wordt verklaard door het positieve teken van $\frac{\partial\eta}{\partial y}$ in combinatie met de richting waarin de drifters zijn gegaan.
